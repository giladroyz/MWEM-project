```julia
d_continuous = Truncated(MixtureModel(Normal, [(.275, .075), (.65, .1)], [0.65, 0.35]), 0, 1)
xs = linspace(d_continuous.lower, d_continuous.upper, 300)
ys = pdf.(d_continuous, xs)
```

```julia
num_samples = 1000
domain_size = [10 10 10]
samples = [rand(d_continuous, num_samples) rand(d_continuous, num_samples) rand(d_continuous, num_samples)]
domain_size
```

```julia
function get_flatten_index(index, dim)
    counter = 0
    for (p,(i,s)) in enumerate(zip(index,dim))
        counter += i*(s^(p-1))
        #println(p)
        #println(i)
        #println(s)
        #println()
    end
    counter
end

function histogram_from_sample(samples::AbstractArray, domain_dim::Tuple)
    @assert 0 <= minimum(samples)
    @assert maximum(samples) <= 1
    ones_array = ones(Int64, domain_dim)
    histogram_size = prod(domain_dim)
    h = zeros(Int64, domain_dim)
    rows, cols = size(samples)
    
    function f(row)
        [row[i]*domain_dim[i] for i=1:length(row)]
    end

    new_samples = [f(s)[1] for s in samples]
    for i =1:rows
        x = round.(Int64, new_samples[i, :])
        index = get_flatten_index(x,domain_dim)
        @inbounds h[index] += 1
    end
    collect(1:histogram_size)/histogram_size, h[:]
end
```

```julia
xs, ys = histogram_from_sample(samples, Tuple(domain_size))
bar(xs, ys, width=1/100);#1/prod(domain_size));
#ys
```
