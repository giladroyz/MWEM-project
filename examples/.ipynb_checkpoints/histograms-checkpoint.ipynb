## Example: Approximating a mixture of two Gaussians.

We start from a continuous mixture of two Gaussians.

```julia
d_continuous = Truncated(MixtureModel(Normal, [(.275, .075), (.65, .1)], [0.65, 0.35]), 0, 1)
xs = linspace(d_continuous.lower, d_continuous.upper, 300)
ys = pdf.(d_continuous, xs)
plot(xs, ys);
```

### Non-private histogram approximation

```julia
num_samples = 1000
domain_size = 100
samples = rand(d_continuous, num_samples)

function histogram_from_sample(samples, num_bins)
    @assert 0 <= minimum(samples)
    @assert maximum(samples) <= 1
    h = zeros(num_bins)
    new_samples = samples*num_bins
    for x in round.(Int64, new_samples)
        h[x] += 1
    end
    collect(1:num_bins)/num_bins, h
end

xs, ys = histogram_from_sample(samples, domain_size)
bar(xs, ys, width=1/domain_size);
```

### Private histogram approximation

Obtaining a private approximation is as easy as running the following command.

```julia
mw = mwem(SeriesRangeQueries(domain_size), Histogram(ys/sum(ys), num_samples));
bar(xs, mw.synthetic.weights*num_samples, align="center", color="red", alpha=0.3, width=1/domain_size);
```

Above we ran `mwem` with the default parameters corresponding to `10` iterations and `epsilon=1.0` per iteration. We can adjust these parameters as follows:

```julia
mw = mwem(SeriesRangeQueries(domain_size), Histogram(ys/sum(ys), num_samples), MWParameters(epsilon=0.1, iterations=10, noisy_init=true));
error1 = mean_squared_error(mw)
#bar(xs, mw.synthetic.weights*num_samples, align="center", color="red", alpha=0.3, width=1/domain_size);
mw1 = mwem(SeriesRangeQueries(domain_size), Histogram(ys/sum(ys), num_samples), MWParameters(epsilon=0.1, iterations=10));
#bar(xs, mw.synthetic.weights*num_samples, align="center", color="red", alpha=0.3, width=1/domain_size);
error2 = mean_squared_error(mw1)

println(error1)
println(error2)
```

### Side-by-side comparison

```julia
bar(xs, ys, 1/domain_size);
bar(xs, mw1.synthetic.weights*num_samples, align="center", color="red", alpha=0.3, width=1/domain_size);
```

```julia
function countmemb(itr)
    d = Dict{eltype(itr), Int}()
    for val in itr
        if isa(val, Number) && isnan(val)
            continue
        end
        d[val] = get!(d, val, 0) + 1
    end
    return d
end
```

```julia
function getLowerBound(queries::HistogramQueries, eps, delta)
    m,n = size(queries.queries)
    P_eps_gamma = (2*log2(2/delta))/(eps^2)
    singular_values = svdfact(queries.queries)[:S]
    ss_svd = (1/n)*(sum(singular_values)^2)

    ss_svd*P_eps_gamma
end
```

```julia
function get_query_vector_from_interval(interval::Interval, domain::Int64)
    query_vector = zeros(domain)
    start_index, end_index = interval
    #print(start_index, end_index)
    query_vector[start_index : end_index] = 1
    query_vector
    #println(query_vector)
end
```
